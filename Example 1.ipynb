```python
class Rectangle():
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def Area(self):
    return (self.length)*(self.width)

  def Display(self):
    print ("The area of the rectangle is", self.Area())

  length = float(input("Input length:"))
  width = float(input("Input width:"))

  shape = Rectangle(length,width)
  shape.Display()
```
